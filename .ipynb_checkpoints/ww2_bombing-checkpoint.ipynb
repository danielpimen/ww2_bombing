```python
import pandas as pd
```

```python
# Import airplane list 
csv_path = "THOR_WWII_AIRCRAFT_GLOSS.csv"
airplane_list = pd.read_csv(csv_path, encoding="utf-8")
airplane_list.head()
```

```python
#import bombing list
csv_path2 = "THOR_WWII_DATA_CLEAN.csv"
bombing_list = pd.read_csv(csv_path2, encoding = "ISO-8859-1")
bombing_list.head()
```

```python
reduced_bombing_df = bombing_list[["WWII_ID", "MSNDATE", "THEATER", "COUNTRY_FLYING_MISSION",
                       "TGT_COUNTRY", "TGT_LOCATION", "TGT_TYPE", "TGT_INDUSTRY", "AIRCRAFT_NAME"]]
reduced_bombing_df.head()
```

```python
##10 Locations with most bombing 
n = 10
reduced_bombing_df['TGT_LOCATION'].value_counts()[:n].index.tolist()
```

```python
##Most common 3 aircraft used 
n = 3
reduced_bombing_df['AIRCRAFT_NAME'].value_counts()[:n].index.tolist()
```

```python
##Most Targeted Countries 
n = 3
reduced_bombing_df['TGT_COUNTRY'].value_counts()[:n].index.tolist()
```

```python
##Total Number of Bombing Runs
reduced_bombing_df.shape[0]
```

```python
#Need to transform date 
reduced_bombing_df['MSNDATE'] = pd.to_datetime(reduced_bombing_df['MSNDATE'], yearfirst=True).dt.strftime("%Y-%m")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Filter the DataFrame down only to those columns to chart
bomb_numbers_df= reduced_bombing_df[["MSNDATE","TGT_COUNTRY"]]

# Set the index to be "State" so they will be used as labels
country_total_df = bomb_numbers_df.set_index("TGT_COUNTRY")

country_total_df.head()
```

```python
# Multi-column frequency count 
count = country_total_df.groupby(['TGT_COUNTRY']).count() 
print(count) 
count.head()

```

```python
df_filtered = count[count['MSNDATE'] >= 1000] 
import matplotlib.pyplot as plt
df_filtered.reset_index(inplace=True)
df_filtered.head(50)
```

```python
##Remove countries under 5k 
df_filtered.drop(df_filtered.loc[df_filtered['MSNDATE'] < 5000].index, inplace=True)
```

```python
df_filtered.rename(columns={"MSNDATE": "# of Bombings"}, inplace = True)
df_filtered.rename(columns={"TGT_COUNTRY": "Target Country"}, inplace = True)
df_filtered.head(50)
```

```python
df_filtered.plot(x = "Target Country", y = "# of Bombings", kind = "bar")
plt.show()
```
